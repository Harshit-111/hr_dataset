#### checking for above assumption

```python
df1 = pd.read_excel("employees_data.xlsx",'Existing employees')
```

```python
df1.shape
```

```python
df1.head()
```

#### a) Time spent in company 

```python
df_existing = df1['time_spend_company'].mean()
df_left = df['time_spend_company'].mean()

print("existing:",df_existing,"\n","left:",df_left)
```

Exiting employees have spent less time than existing employees

#### b) Average monthly hours 

```python
df_existing = df1['average_montly_hours'].mean()
df_left = df['average_montly_hours'].mean()

print("existing:",df_existing,"\n","left:",df_left)
```

Exiting employee spent more average monthly hours in the company, maybe they were not appreciated or they work in crucial departments

#### c) Number of Projects 

```python
df_existing = df1['number_project'].mean()
df_left = df['number_project'].mean()

print("existing:",df_existing,"\n","left:",df_left)
```

Employees who left had more number of projects than those who are still in the company

#### 3)There is a high correlation between the number of projects and time spent in the company. There is a very high correlation between the number of projects and average number of projects and average monthly hours 

#### 4) There is a high correlation berween the average monthly hours and the time spent with the company (0.66). The more time spent with the company, the higher the average monthly hours and vice versa.

#### 5) Work accidents and promotion in last 5 years has little correlation with other features.  

```python
df_work_accident_left = df.groupby('Work_accident').size().reset_index(name = 'Work Accidents')
df_work_accident_left
```

only 169 out of 3571 exiting employees have had accident at work

```python
df_work_promotion_left = df.groupby('promotion_last_5years').size().reset_index(name = 'had_promotion')
df_work_promotion_left
```

only 19 of 3571 employees have been promoted int he last five years.

### Comparitive Analysis 

```python
df_evaluation = df.groupby('last_evaluation').size().reset_index(name='numbers')
df_evaluation.head()
```

```python
plt.figure(figsize = (8,5))
plt.xlabel("Evalutaion score")
plt.ylabel("Number of Employees")
x=df_evaluation['last_evaluation']
y=df_evaluation['numbers']
plt.bar(x,y)
plt.grid(which = 'both', linestyle = ':')
```

```python
df_projects = df.groupby('number_project').size().reset_index(name='projects')
df_projects.head()
```

```python
plt.figure(figsize = (8,5))
plt.xlabel("number of Projects")
plt.ylabel("Number of Employees")
x=df_projects['number_project']
y=df_projects['projects']
plt.bar(x,y)
plt.grid(which = 'both', linestyle = ':')
```
